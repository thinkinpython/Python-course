## Theoretical recap (quick reminder)

- Python executes code **line by line**, from top to bottom.
- `print()` is used to **display information** on the screen.
- Comments start with `#` and are **ignored by Python**.
- Variables store values using the assignment operator `=`.
- Common basic types are: `int`, `float`, `str`, `bool`, `None`.
- Mathematical operations follow standard rules:
  - `+ - * / // % **`
  - `/` always returns a float
- Strings represent text:
  - written inside quotes
  - support concatenation (`+`) and repetition (`*`)
  - `len()` returns their length
  - indexing starts at 0
- `input()` allows the user to enter data and **always returns a string**.
- Type conversion is done with `int()`, `float()`, `str()`, `bool()`.

### Exercise 5 — Integer division and remainder

Write a program that:

- prints the integer division result
- prints the remainder


```python
a = input("Enter the first integer: ")
b = input("Enter the second integer: ")

a = int(a)
b = int(b)

print(a // b)
print(a % b)
```

### Exercise 6 — Simple price calculation

Write a program that:
- asks the user for a price
- converts the input to a float
- prints the price with VAT (VAT = 21%)


```python
price = input("Enter the price: ")
price = float(price)

price_with_vat = price * 1.21

print(price_with_vat)
```

### Exercise 7 — Final combined exercise

Write a program that:
- asks the user for their name
- asks the user for a number
- converts the number to an integer
- prints the square of the number
- prints a sentence including the user's name and the result


```python
name = input("Enter your name: ")
number = input("Enter a number: ")

number = int(number)
square = number ** 2

print(square)
print(name, "your number squared is", square)
```
