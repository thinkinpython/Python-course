## Theoretical recap (quick reminder)

- Python executes code **line by line**, from top to bottom.
- `print()` is used to **display information** on the screen.
- Comments start with `#` and are **ignored by Python**.
- Variables store values using the assignment operator `=`.
- Common basic types are: `int`, `float`, `str`, `bool`, `None`.
- Mathematical operations follow standard rules:
  - `+ - * / // % **`
  - `/` always returns a float
- Strings represent text:
  - written inside quotes
  - support concatenation (`+`) and repetition (`*`)
  - `len()` returns their length
  - indexing starts at 0
- `input()` allows the user to enter data and **always returns a string**.
- Type conversion is done with `int()`, `float()`, `str()`, `bool()`.
