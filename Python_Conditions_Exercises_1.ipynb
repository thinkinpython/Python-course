# Exercises - Conditions in Python (Beginner)

This notebook contains a short theoretical recap, followed by exercises.

## Theoretical recap (quick reminder)

- A condition checks if something is **True** or **False**.
- `if`, `elif`, `else` let you **execute different blocks** depending on the result.
- Comparison operators: `==`, `!=`, `<`, `>`, `<=`, `>=`
- Logical operators: `and`, `or`, `not`
- Indentation (**4 spaces**) is mandatory inside each block.
- Conditions can be **nested** inside other conditions.
- `if` is mandatory, `elif` and `else` are optional.

### Exercise 1 — Positive, negative or zero

Write a program that:
- asks the user for a number
- converts it to a float
- prints whether the number is **positive**, **negative**, or **zero**

```python
number = float(input("Enter a number: "))

if number > 0:
    print("The number is positive")
elif number < 0:
    print("The number is negative")
else:
    print("The number is zero")
```

### Exercise 2 — Even or odd

Write a program that:
- asks the user for an integer
- prints whether the number is **even** or **odd**

```python
number = int(input("Enter an integer: "))

if number % 2 == 0:
    print("The number is even")
else:
    print("The number is odd")
```

### Exercise 3 — Access control

Write a program that:
- asks the user for their age
- prints `"Access granted"` if they are **18 or older**
- prints `"Access denied"` otherwise

```python
age = int(input("Enter your age: "))

if age >= 18:
    print("Access granted")
else:
    print("Access denied")
```

### Exercise 4 — Grade classification

Write a program that:
- asks the user for a grade between 0 and 100
- prints:
  - `"Excellent"` if the grade is **90 or above**
  - `"Good"` if the grade is between **70 and 89**
  - `"Average"` if the grade is between **50 and 69**
  - `"Failing"` if the grade is **below 50**

```python
grade = float(input("Enter your grade (0-100): "))

if grade >= 90:
    print("Excellent")
elif grade >= 70:
    print("Good")
elif grade >= 50:
    print("Average")
else:
    print("Failing")
```

### Exercise 5 — Login system

Write a program that:
- stores a username and a password in two variables (you choose the values)
- asks the user to enter a username and a password
- prints `"Login successful"` only if **both** are correct
- prints `"Wrong username or password"` otherwise

```python
correct_username = "admin"
correct_password = "1234"

username = input("Enter your username: ")
password = input("Enter your password: ")

if username == correct_username and password == correct_password:
    print("Login successful")
else:
    print("Wrong username or password")
```

### Exercise 6 — Biggest of three numbers

Write a program that:
- asks the user for **three numbers**
- converts them to floats
- prints which one is the **biggest**

```python
a = float(input("Enter the first number: "))
b = float(input("Enter the second number: "))
c = float(input("Enter the third number: "))

if a >= b and a >= c:
    print("The biggest number is", a)
elif b >= a and b >= c:
    print("The biggest number is", b)
else:
    print("The biggest number is", c)
```

### Exercise 7 — Small combined exercise

Write a program that:
- asks the user for their **name** and **age**
- prints:
  - `"{name} is a child"` if the user is **under 13**
  - `"{name} is a teenager"` if the user is **between 13 and 17**
  - `"{name} is an adult"` if the user is **18 or older**
- in **all cases**, also prints `"Welcome, {name}!"`

```python
name = input("Enter your name: ")
age = int(input("Enter your age: "))

if age < 13:
    print(name, "is a child")
elif age < 18:
    print(name, "is a teenager")
else:
    print(name, "is an adult")

print("Welcome,", name + "!")
```
