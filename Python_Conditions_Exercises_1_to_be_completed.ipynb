# Exercises - Conditions in Python (Beginner)

This notebook contains a short theoretical recap, followed by exercises.

## Theoretical recap (quick reminder)

- A condition checks if something is **True** or **False**.
- `if`, `elif`, `else` let you **execute different blocks** depending on the result.
- Comparison operators: `==`, `!=`, `<`, `>`, `<=`, `>=`
- Logical operators: `and`, `or`, `not`
- Indentation (**4 spaces**) is mandatory inside each block.
- Conditions can be **nested** inside other conditions.
- `if` is mandatory, `elif` and `else` are optional.
