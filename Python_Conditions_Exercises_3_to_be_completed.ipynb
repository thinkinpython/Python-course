# Exercises - Conditions in Python (Advanced)

This notebook contains 3 advanced exercises that combine conditions with
string manipulation, nested logic, and multiple validation rules.

## Theoretical recap (quick reminder)

- `str.isdigit()` → returns `True` if all characters are digits
- `str.isalpha()` → returns `True` if all characters are letters
- `str.islower()` / `str.isupper()` → checks the case of a string
- `str.lower()` / `str.upper()` → converts a string to lower/upper case
- `in` works on strings: `"hello" in "say hello world"` → `True`
- You can chain comparisons: `0 <= x <= 100` is valid in Python
- `abs()` returns the absolute value of a number
- A condition stored in a variable is just a `bool`: `is_valid = len(password) >= 8`

### Exercise 1 — Advanced password validator

Write a program that checks if a password is **strong**.

The password must meet **all** of the following conditions:
- At least **10 characters** long
- Contains at least one **uppercase letter** (hint: check each character with a loop... or think differently)
- Contains at least one **digit**
- Contains at least one of these special characters: `@`, `#`, `$`, `!`
- Does **not** contain any spaces

For each broken rule, print a **specific error message**.
If all rules are met, print `"Strong password ✓"`

> 💡 Hint: you can use `any()` combined with a generator expression.
> For example: `any(c.isupper() for c in password)` returns `True`
> if at least one character is uppercase.

### Exercise 2 — Train ticket pricing system

A train company applies the following pricing rules:

**Base price: 50€**

Discounts and surcharges are applied **on top of the base price**:
- **Child** (under 12): −50%
- **Senior** (65 or older): −30%
- **Adult** (18–64): no discount
- **Teen** (12–17): −10%

Then, depending on the travel class:
- **First class**: +40% on the current price
- **Second class**: no change

Finally:
- If the user has a **loyalty card** (`"yes"` / `"no"`): additional −15%

Write a program that:
- asks for the passenger's age, travel class (`1` or `2`), and loyalty card status
- computes and prints the **final ticket price**, rounded to 2 decimal places
- prints a **summary** of all discounts or surcharges applied

### Exercise 3 — Credit card number validator

Write a program that validates a credit card number entered as a string.

The program must check **all** of the following rules:
- The number contains **exactly 16 characters**
- The number contains **only digits** (no spaces, no dashes)
- The number does **not** start with `0`
- The **sum of all digits** is divisible by 10 *(this is a simplified version of the Luhn algorithm)*

For each broken rule, print a **specific error message**.
If all rules pass, print `"Card number is valid ✓"`

> 💡 Hint: to sum the digits, you can use:
> `total = sum(int(c) for c in number)`
