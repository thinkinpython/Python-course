# Exercises - Conditions in Python (Intermediate)

This notebook contains a short theoretical recap, followed by intermediate exercises on conditions.

## Theoretical recap (quick reminder)

- Conditions can be **combined** with `and`, `or`, `not`.
- You can **nest** an `if` inside another `if`.
- `in` checks if a value is **inside a string or a list**: `"a" in "cat"` → `True`
- A condition can be stored in a variable: `is_adult = age >= 18`
- Be careful with **operator priority**: `not` > `and` > `or`
- An `if` block can contain **as many lines as needed**, not just one.

### Exercise 1 — Ticket price

A cinema applies the following pricing rules:
- Under 12 years old → **5€**
- Between 12 and 17 years old → **8€**
- Between 18 and 64 years old → **12€**
- 65 years old or older → **7€**

Write a program that:
- asks the user for their age
- prints the correct ticket price
- prints a sentence like: `"Your ticket costs 12€"`

```python
age = int(input("Enter your age: "))

if age < 12:
    price = 5
elif age < 18:
    price = 8
elif age < 65:
    price = 12
else:
    price = 7

print("Your ticket costs " + str(price) + "€")
```

### Exercise 2 — Triangle type

Write a program that:
- asks the user for **three side lengths** of a triangle (as floats)
- prints whether the triangle is:
  - `"Equilateral"` → all three sides are equal
  - `"Isosceles"` → exactly two sides are equal
  - `"Scalene"` → all three sides are different

> 💡 Hint: use `and` and `or` to combine your conditions.

```python
a = float(input("Enter the first side: "))
b = float(input("Enter the second side: "))
c = float(input("Enter the third side: "))

if a == b and b == c:
    print("Equilateral")
elif a == b or b == c or a == c:
    print("Isosceles")
else:
    print("Scalene")
```

### Exercise 3 — Password validator

Write a program that checks if a password is valid.

The password must meet **all** of the following conditions:
- At least **8 characters** long
- Contains the character `"@"`
- Does not start with a number

If all conditions are met → print `"Password accepted"`
Otherwise → print a **specific message** for each rule that is broken.

> 💡 Hint: use `len()` to check the length, `in` to check for a character,
> and `str.isdigit()` on the first character.

```python
password = input("Enter a password: ")

is_valid = True

if len(password) < 8:
    print("Error: password must be at least 8 characters long")
    is_valid = False

if "@" not in password:
    print("Error: password must contain '@'")
    is_valid = False

if password[0].isdigit():
    print("Error: password must not start with a number")
    is_valid = False

if is_valid:
    print("Password accepted")
```

### Exercise 4 — BMI calculator

The Body Mass Index (BMI) is calculated as follows:

**BMI = weight (kg) / height (m)²**

Write a program that:
- asks the user for their weight in kg and height in meters
- computes their BMI (rounded to 1 decimal place with `round()`)
- prints the BMI value
- prints the corresponding category:
  - BMI **below 18.5** → `"Underweight"`
  - BMI between **18.5 and 24.9** → `"Normal weight"`
  - BMI between **25 and 29.9** → `"Overweight"`
  - BMI **30 or above** → `"Obese"`

```python
weight = float(input("Enter your weight in kg: "))
height = float(input("Enter your height in meters: "))

bmi = round(weight / height ** 2, 1)

print("Your BMI is", bmi)

if bmi < 18.5:
    print("Category: Underweight")
elif bmi < 25:
    print("Category: Normal weight")
elif bmi < 30:
    print("Category: Overweight")
else:
    print("Category: Obese")
```

### Exercise 5 — Small combined exercise

Write a program that simulates a **simple number guessing game** (without a loop).

- Store a secret number in a variable (you choose it)
- Ask the user to guess the number
- Convert the input to an integer
- Print one of the following:
  - `"Too low! But you're close"` if the guess is below the secret number by **5 or less**
  - `"Too low!"` if the guess is below the secret number by **more than 5**
  - `"Too high! But you're close"` if the guess is above the secret number by **5 or less**
  - `"Too high!"` if the guess is above the secret number by **more than 5**
  - `"Correct! Well done."` if the guess is exactly right

> 💡 Hint: compute the **difference** between the guess and the secret number first.

```python
secret = 42

guess = int(input("Guess the secret number: "))

difference = secret - guess

if guess == secret:
    print("Correct! Well done.")
elif difference > 0 and difference <= 5:
    print("Too low! But you're close")
elif difference > 5:
    print("Too low!")
elif difference < 0 and difference >= -5:
    print("Too high! But you're close")
else:
    print("Too high!")
```
